# NEURAL NETWORK UNSCALED

```python
# Import necessary libraries
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier

# Function to perform cross-validation and find the best hidden layer size
def find_best_hidden_layer_size(hidden_layer_sizes, x_train, y_train):
    best_score = 0
    best_size = None
    for size in hidden_layer_sizes:
        mlp = MLPClassifier(hidden_layer_sizes=size, max_iter=3000, random_state=42)
        scores = cross_val_score(mlp, x_train, y_train, cv=5, scoring='accuracy')
        mean_score = scores.mean()
        print(f'Hidden layer size {size}: Mean cross-validation accuracy = {mean_score}')
        if mean_score > best_score:
            best_score = mean_score
            best_size = size
    return best_size, best_score

# Define hidden layer sizes to test
hidden_layer_sizes_1 = [(h,) for h in [5] + list(range(10, 101, 5))]
hidden_layer_sizes_2 = [(5, h) for h in [5] + list(range(10, 101, 5))]

# Find the best hidden layer size for 1 hidden layer
best_size_1, best_score_1 = find_best_hidden_layer_size(hidden_layer_sizes_1, x_train, y_train)
print(f'Best hidden layer size for 1 layer: {best_size_1} with score {best_score_1}')

# Find the best hidden layer size for 2 hidden layers
best_size_2, best_score_2 = find_best_hidden_layer_size(hidden_layer_sizes_2, x_train, y_train)
print(f'Best hidden layer size for 2 layers: {best_size_2} with score {best_score_2}')
```

```python
# Train the final model with the best configuration and evaluate it
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

def train_and_evaluate(hidden_layer_size, x_train, y_train, x_test, y_test):
    mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_size, max_iter=3000, random_state=42)
    mlp.fit(x_train, y_train)
    test_predict = mlp.predict(x_test)
    accuracy = metrics.accuracy_score(y_test, test_predict)
    print(f'Accuracy (Test) for hidden layer size {hidden_layer_size}: {accuracy}')
    print(confusion_matrix(y_test, test_predict))
    print(classification_report(y_test, test_predict))
```

```python
# Evaluate the best models
print("Evaluation of the best model with 1 hidden layer:")
train_and_evaluate(best_size_1, x_train, y_train, x_test, y_test)
```

```python
# Evaluate the best models
print("Evaluation of the best model with 2 hidden layers:")
train_and_evaluate(best_size_2, x_train, y_train, x_test, y_test)
```

# SVM UNSCALED

```python
# Import necessary libraries
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Define the parameter range for C
C_range = [0.1,1,10,100]
degree = [1,2,3,4,5]
```

LINEAR SVM

```python
param_grid_linear = {'C': C_range}

# Function to perform Grid Search and evaluate the model for linear kernel
def evaluate_linear_kernel(x_train, y_train, x_test, y_test):
    print("Evaluating linear kernel")
    grid_search = GridSearchCV(SVC(kernel='linear'), param_grid_linear, cv=5, scoring='accuracy')
    grid_result = grid_search.fit(x_train, y_train)
    print(f"Best parameters for linear kernel: {grid_result.best_params_}")
    
    # Train the model with the best parameters
    best_model = grid_result.best_estimator_
    best_model.fit(x_train, y_train)
    
    # Predict on the test set
    test_predict = best_model.predict(x_test)
    
    # Compute and print the model accuracy on the test set
    accuracy = metrics.accuracy_score(y_test, test_predict)
    print(f"Accuracy (Test) for linear kernel: {accuracy}")
    
    # Print confusion matrix and classification report
    print(confusion_matrix(y_test, test_predict))
    print(classification_report(y_test, test_predict))
    print("\n" + "="*60 + "\n")

#print the best evaluate linear kernel    
evaluate_linear_kernel(x_train, y_train, x_test, y_test)
```
