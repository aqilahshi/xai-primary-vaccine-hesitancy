```python
target=df["C.Non_hesitant"]
features=df.drop('C.Non_hesitant',axis=1)
```

# SPLIT UNSCALED

```python
# Split data into train and test sets

# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split the dataset into training + development set and test set
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
```

# KNN UNSCALED

```python
from sklearn.neighbors import KNeighborsClassifier
# Train the model using the entire training set with the best k
knn_best = KNeighborsClassifier(n_neighbors=9)
knn_best.fit(x_train, y_train)
```

```python
# Predict the target for the test dataset
test_predict = knn_best.predict(x_test)
test_predict
```

```python
# Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test, test_predict)
print("Accuracy: ", accuracy)
```

```python
# Import scikit-learn metrics methods
from sklearn.metrics import classification_report, confusion_matrix

# Print confusion matrix
print(confusion_matrix(y_test, test_predict))

# Print precision, recall and F1 scores
print(classification_report(y_test, test_predict))
```

# DECISION TREE UNSCALED 

```python
from sklearn.tree import DecisionTreeClassifier
# Train the model using the entire training set with the best depth
dtree_best = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
dtree_best.fit(x_train, y_train)
```

```python
# Predict the target for the test dataset
test_predict = dtree_best.predict(x_test)

test_predict
```

```python
# Evaluate the decision tree model on test set

# Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Compute the model accuracy on the test set: How often is the classifier correct?
accuracy = metrics.accuracy_score(y_test, test_predict)
print("Accuracy: ", accuracy)
```

```python
# Import scikit-learn metrics methods
from sklearn.metrics import classification_report, confusion_matrix

# Print confusion matrix
print(confusion_matrix(y_test, test_predict))

# Print precision, recall and F1 scores
print(classification_report(y_test, test_predict))
```

# BAYES UNSCALED

Multinomial Naive Bayes

```python
# Import the Multinomial Naive Bayes classifier
from sklearn.naive_bayes import MultinomialNB
# Import function for k-fold cross validation
from sklearn.model_selection import cross_val_score

# Create a Multinomial Naive Bayes classifier with default parameters
mnb = MultinomialNB()

# Use 5-fold cross validation to perform training and validation on the training set
scores = cross_val_score(mnb, x_train, y_train, cv = 5, scoring = 'accuracy')

# Display the array containing accuracy from 10 folds or iterations
scores
```

```python
# Print the mean accuracy score
print('Accuracy (Validation) =', scores.mean())
```

```python
# Running prediction
mnb.fit(x_train, y_train)

# Predict the target for the test dataset
test_predict = mnb.predict(x_test)

# Compute the model accuracy on the development set: How often is the classifier correct?
print("Accuracy (Test): ", metrics.accuracy_score(y_test, test_predict))
```

```python
# Import scikit-learn metrics methods
from sklearn.metrics import classification_report, confusion_matrix

# Print confusion matrix
print(confusion_matrix(y_test, test_predict))

# Print precision, recall and F1 scores
print(classification_report(y_test, test_predict))
```

# NEURAL NETWORK UNSCALED

```python
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Define the hidden layer sizes
hidden_layer_size = (5,10)

def train_and_evaluate(hidden_layer_size, x_train, y_train, x_test, y_test):
    # Initialize the MLPClassifier with the given hidden layer size
    mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_size, max_iter=3000, random_state=42)
    # Train the model
    mlp.fit(x_train, y_train)
    # Make predictions
    test_predict = mlp.predict(x_test)
    # Evaluate the model
    accuracy = metrics.accuracy_score(y_test, test_predict)
    print(f'Accuracy (Test) for hidden layer size {hidden_layer_size}: {accuracy}')
    print(confusion_matrix(y_test, test_predict))
    print(classification_report(y_test, test_predict))
```

```python
# Call the function with the actual data
train_and_evaluate(hidden_layer_size, x_train, y_train, x_test, y_test)
```

# SVM UNSCALED

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Linear SVM
print("Evaluating Linear SVM")
model_linear = SVC(kernel='linear', C=0.1)
model_linear.fit(x_train, y_train)

# Predict and evaluate
y_pred_linear = model_linear.predict(x_test)
acc_linear = metrics.accuracy_score(y_test, y_pred_linear)
print(f"Accuracy (Test) for Linear SVM: {acc_linear}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_linear))
print("Classification Report:")
print(classification_report(y_test, y_pred_linear))
print("\n" + "="*60 + "\n")
```

```python
# RBF SVM
print("Evaluating RBF SVM")
model_rbf = SVC(kernel='rbf', C=10, gamma='scale')
model_rbf.fit(x_train, y_train)

# Predict and evaluate
y_pred_rbf = model_rbf.predict(x_test)
acc_rbf = metrics.accuracy_score(y_test, y_pred_rbf)
print(f"Accuracy (Test) for RBF SVM: {acc_rbf}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rbf))
print("Classification Report:")
print(classification_report(y_test, y_pred_rbf))
print("\n" + "="*60 + "\n")
```
