```python
import pandas as pd
df=pd.read_csv('primary_input.csv')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df
```

```python
import warnings
import numpy as np

# Suppress overflow warning
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(over='ignore')
```

```python
import numpy as np

# Apply the threshold to create the hesitant column
df['C.Non_hesitant'] = np.where(df['C.target_mean'] > 2.8, 1, 0)
# Display the resulting DataFrame with the aggregated target and hesitant column
print(df[['C.target_mean', 'C.Non_hesitant']])
```

```python
df.drop(columns ='C.target_mean', inplace=True)
df.head()
```

```python
target=df["C.Non_hesitant"]
features=df.drop('C.Non_hesitant',axis=1)
```

```python
# Split data into train and test sets

# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split the dataset into training + development set and test set
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
```

# RANDOM FOREST

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# Random Forest Bagging
model_bagging_rf = BaggingClassifier(estimator=RandomForestClassifier(), n_estimators=10, random_state=7)

# Cross-validation
print("Model: Random Forest")
scores_rf = cross_val_score(model_bagging_rf, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_rf)
print("Mean Accuracy (Validation):", scores_rf.mean())
print("-" * 50)

# Test set evaluation
model_bagging_rf.fit(x_train, y_train)
test_predict_rf = model_bagging_rf.predict(x_test)
print("Random Forest Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_rf))
print("=" * 50)
```

# BAGGING

```python
# Import necessary libraries
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# Neural Network with one hidden layer of 5 neurons
model_bagging_nn = BaggingClassifier(
    estimator=MLPClassifier(hidden_layer_sizes=(5,10), max_iter=200, random_state=7), 
    n_estimators=10, 
    random_state=7
)

# Cross-validation
print("Model: Neural Network (5 neurons in the hidden layer)")
scores_nn = cross_val_score(model_bagging_nn, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold: ", scores_nn)
print("Mean Accuracy (Validation):", scores_nn.mean())
print("-" * 50)

# Test set evaluation
model_bagging_nn.fit(x_train, y_train)
test_predict_nn = model_bagging_nn.predict(x_test)
print("Neural Network Bagging - Accuracy (Test):", metrics.accuracy_score(y_test, test_predict_nn))
print("=" * 50)
```

# Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# Define the Gradient Boosting classifier
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Cross-validation
cv_scores = cross_val_score(gbc, x_train, y_train, cv=5, scoring='accuracy')
print("Accuracy scores for each fold (CV): ", cv_scores)
print("Mean Accuracy (Validation): {:.4f}".format(cv_scores.mean()))
print("-" * 50)

# Fit the model on the full training data
gbc.fit(x_train, y_train)

# Evaluate the model on the test data
y_test_pred = gbc.predict(x_test)
accuracy = accuracy_score(y_test, y_test_pred)

# Print evaluation metrics
print(f'Accuracy (Test): {accuracy:.4f}')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_test_pred))
print('Classification Report:')
print(classification_report(y_test, y_test_pred))
print("=" * 50)
```

```python
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
from sklearn import metrics

# KNN (k=23) - Bagging only
model_bagging_knn = BaggingClassifier(
    estimator=KNeighborsClassifier(n_neighbors=9),
    n_estimators=10,
    random_state=7
)
print("Model: KNN (k=23) - Bagging")
scores = cross_val_score(model_bagging_knn, x_train, y_train, cv=5, scoring='accuracy')
print("Validation Accuracy:", scores.mean())
model_bagging_knn.fit(x_train, y_train)
print("Test Accuracy:", metrics.accuracy_score(y_test, model_bagging_knn.predict(x_test)))
print("=" * 50)

# Decision Tree (depth=4) - Bagging
model_bagging_dt = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, random_state=7),
    n_estimators=10,
    random_state=7
)
print("Model: Decision Tree (depth=4) - Bagging")
scores = cross_val_score(model_bagging_dt, x_train, y_train, cv=5, scoring='accuracy')
print("Validation Accuracy:", scores.mean())
model_bagging_dt.fit(x_train, y_train)
print("Test Accuracy:", metrics.accuracy_score(y_test, model_bagging_dt.predict(x_test)))
print("=" * 50)

# Decision Tree (depth=4) - AdaBoost
model_adaboost_dt = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, random_state=7),
    n_estimators=50,
    random_state=7
)
print("Model: Decision Tree (depth=3) - AdaBoost")
scores = cross_val_score(model_adaboost_dt, x_train, y_train, cv=5, scoring='accuracy')
print("Validation Accuracy:", scores.mean())
model_adaboost_dt.fit(x_train, y_train)
print("Test Accuracy:", metrics.accuracy_score(y_test, model_adaboost_dt.predict(x_test)))
print("=" * 50)

# MultinomialNB - Bagging only
model_bagging_mnb = BaggingClassifier(
    estimator=MultinomialNB(),
    n_estimators=10,
    random_state=7
)
print("Model: MultinomialNB - Bagging")
scores = cross_val_score(model_bagging_mnb, x_train, y_train, cv=5, scoring='accuracy')
print("Validation Accuracy:", scores.mean())
model_bagging_mnb.fit(x_train, y_train)
print("Test Accuracy:", metrics.accuracy_score(y_test, model_bagging_mnb.predict(x_test)))
print("=" * 50)
```
